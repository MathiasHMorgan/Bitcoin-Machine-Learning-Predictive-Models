```python
df = btc_daily_df_adjusted 
```

```python
prices = df['Close']
returns = prices.pct_change()
```

```python
prices.iloc[0]
```

```python
prices.iloc[800]
```

```python
last_price = prices.iloc[800]
```

```python
num_simulations = 1000
num_days = 200
```

```python
simulation_df = {}
```

```python
 daily_vol = returns.std()

for x in range(num_simulations):
    count = 0
   
    price_series = []

    price = last_price * (1 + np.random.normal(0, daily_vol))
    price_series.append(price)

    for y in range(num_days - 1):
        price = price_series[count] * (1 + np.random.normal(0, daily_vol))
        price_series.append(price)
        count += 1

    simulation_df[str(x)] = price_series
```

```python
simulation_df
```

```python
simulation_df = pd.DataFrame(simulation_df)
```

```python
fig = plt.figure()
fig.suptitle('Monte Carlo Simulation: BTC')
plt.plot(simulation_df)
plt.axhline(y = last_price, color = 'r', linestyle = '_')
plt.xlabel('Day')
plt.ylabel('Price')
plt.show()
```
